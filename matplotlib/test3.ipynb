```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fr = pd.read_csv("./students.csv",delimiter=";",names=["prep", "group", "mark"])
fr_cp = fr.copy()
fr_cp["marks_data"] = fr["mark"]
fr_cp
```

```python

df_pivot = pd.pivot_table(
    fr_cp,
    values="marks_data",
    index="prep",
    columns="mark",
    aggfunc="count"
)
print(df_pivot)
df_pivot.plot.bar(stacked=True,legend='reverse')
```

```python
# Пироги по группам
f_pivot = pd.pivot_table(
    fr_cp,
    values="marks_data",
    index="group",
    columns="mark",
    aggfunc="count"
)
for gr in df_pivot.index:
    ax = df_pivot.loc[gr].plot.pie(subplots=False)
    plt.show()
# handles, labels = ax.get_legend_handles_labels()
# ax.legend(reversed(handles), reversed(labels), loc='upper right')
```

```python
#Столбчатая по группам
f_pivot = pd.pivot_table(
    fr_cp,
    values="marks_data",
    index="group",
    columns="mark",
    aggfunc="count"
)
ax=df_pivot.plot.bar(stacked=True,legend='reverse')
from matplotlib.ticker import MaxNLocator
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
handles, labels = ax.get_legend_handles_labels()
ax.legend(reversed(handles), reversed(labels), loc='upper right')
```
