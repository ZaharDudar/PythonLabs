```python
import matplotlib.pyplot as plt
import numpy as np 
data=[]
with open("./signals/signal01.dat", "r") as f:
    data = [float(i.strip()) for i in f.readlines()]
data = np.array(data)
data
```

```python
def get_av(data,N):
    cumsum = np.cumsum(data)
    print(cumsum[N:]-cumsum[:-N])
    return (cumsum[N:] - cumsum[:-N])/N
data = get_av(data, 10)
data
```

```python
plt.plot(data)
```
