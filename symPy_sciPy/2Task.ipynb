```python
import numpy as np
import matplotlib.pyplot as plt
raw_matrix = []
with open("./large.txt", "r") as f:
    N = int(f.readline().strip())
    lines = f.readlines()
    for line in lines[:-1]:
        raw_matrix.append([float(i) for i in line.strip().split()])
    b = np.array([float(i) for i in lines[-1].strip().split()])
    # b.resize((N,1))
matrix=np.matrix(raw_matrix)
print(matrix)
print(b)
```

```python
solution = np.linalg.solve(matrix, b)
solution.tolist()
```

```python
plt.bar(range(N), solution)
plt.grid()
```
